```python
import ast
import builtins
from dataclasses import dataclass
from typing import Literal, Any

_BUILTINS = set(dir(builtins))

@dataclass
class CanonResult:
    ok: bool
    rep: Any                # either str (canon_str) or list[str] (node_types)
    err: str | None = None  # error message if ok=False


class Canonicalizer(ast.NodeTransformer):
    def __init__(self):
        super().__init__()
        self.name_map: dict[str, str] = {}
        self.counter = 0

    def _map_name(self, s: str, prefix: str = "VAR") -> str:
        if s in _BUILTINS:
            return s
        if s not in self.name_map:
            self.name_map[s] = f"{prefix}_{self.counter}"
            self.counter += 1
        return self.name_map[s]

    def _strip_docstring_from_body(self, body):
        if not body:
            return body
        first = body[0]
        if isinstance(first, ast.Expr) and isinstance(getattr(first, "value", None), ast.Constant):
            if isinstance(first.value.value, str):
                return body[1:]
        return body

    def visit_Module(self, node: ast.Module):
        node = self.generic_visit(node)
        node.body = self._strip_docstring_from_body(node.body)
        return node

    def visit_FunctionDef(self, node: ast.FunctionDef):
        node = self.generic_visit(node)
        node.body = self._strip_docstring_from_body(node.body)
        return node

    def visit_AsyncFunctionDef(self, node: ast.AsyncFunctionDef):
        node = self.generic_visit(node)
        node.body = self._strip_docstring_from_body(node.body)
        return node

    def visit_ClassDef(self, node: ast.ClassDef):
        node = self.generic_visit(node)
        node.body = self._strip_docstring_from_body(node.body)
        return node

    def visit_arg(self, node: ast.arg):
        node.arg = self._map_name(node.arg, prefix="ARG")
        return node

    def visit_Name(self, node: ast.Name):
        node.id = self._map_name(node.id, prefix="VAR")
        return node

    def visit_Constant(self, node: ast.Constant):
        v = node.value
        if isinstance(v, (int, float, complex)):
            node.value = "NUM"
        elif isinstance(v, str):
            node.value = "STR"
        elif isinstance(v, bytes):
            node.value = "BYTES"
        elif v is None:
            node.value = "NONE"
        elif isinstance(v, bool):
            node.value = "BOOL"
        else:
            node.value = "CONST"
        return node


def canonicalize(
    code: str,
    rep: Literal["dump", "node_types"] = "dump",
) -> CanonResult:
    """
    rep="dump"      -> returns ast.dump(canonical_tree, ...) as a string
    rep="node_types"-> returns list[str] of node type names (from ast.walk) on canonical_tree
    """
    try:
        tree = ast.parse(code)
        canon_tree = Canonicalizer().visit(tree)
        ast.fix_missing_locations(canon_tree)

        if rep == "dump":
            canon_str = ast.dump(canon_tree, annotate_fields=False, include_attributes=False)
            return CanonResult(ok=True, rep=canon_str, err=None)

        if rep == "node_types":
            seq = [type(n).__name__ for n in ast.walk(canon_tree)]
            return CanonResult(ok=True, rep=seq, err=None)

        return CanonResult(ok=False, rep="" if rep == "dump" else [], err=f"Unknown rep: {rep}")

    except SyntaxError as e:
        return CanonResult(ok=False, rep="" if rep == "dump" else [], err=f"SyntaxError: {e.msg}")
    except Exception as e:
        return CanonResult(ok=False, rep="" if rep == "dump" else [], err=f"{type(e).__name__}: {e}")
```

```python
import re

def check_indent(code: str):
    lines = code.splitlines()
    stats = {"tab": 0, "space": 0, "mixed": 0, "none": 0}
    examples = {"tab": [], "space": [], "mixed": []}

    for i, line in enumerate(lines, start=1):
        if not line.strip():
            continue  # ignore blank lines

        m = re.match(r"^([ \t]+)", line)
        if not m:
            stats["none"] += 1
            continue

        ws = m.group(1)
        has_tab = "\t" in ws
        has_space = " " in ws

        if has_tab and has_space:
            stats["mixed"] += 1
            if len(examples["mixed"]) < 5:
                examples["mixed"].append((i, repr(ws), line.lstrip()[:80]))
        elif has_tab:
            stats["tab"] += 1
            if len(examples["tab"]) < 5:
                examples["tab"].append((i, repr(ws), line.lstrip()[:80]))
        else:
            stats["space"] += 1
            if len(examples["space"]) < 5:
                examples["space"].append((i, repr(ws), line.lstrip()[:80]))

    return stats, examples


# Example usage:
code = """PASTE_CODE_HERE"""
stats, examples = check_indent(code)
print("Indent stats:", stats)
print("\nExamples (line, leading_ws, snippet):")
for k in ["mixed", "tab", "space"]:
    for ex in examples[k]:
        print(k, ex)

def tabs_to_spaces(code: str, tabsize: int = 4) -> str:
    return "\n".join(line.expandtabs(tabsize) for line in code.splitlines())


code = rows[1]['code_ref']
ast.parse(tabs_to_spaces(code))
```

```python
import hashlib
import struct
from typing import List

import hashlib, struct

I64_MAX = (1 << 63) - 1

def _h63(s: str, seed: int) -> int:
    h = hashlib.blake2b(digest_size=8, person=seed.to_bytes(8, "little"))
    h.update(s.encode("utf-8", errors="ignore"))
    hv = struct.unpack("<Q", h.digest())[0]
    return hv & I64_MAX  # force into signed int64 range

def shingles_from_node_types(node_types: List[str], n: int = 5) -> set[str]:
    if not node_types or len(node_types) < n:
        return set()
    # token n-grams; join with separator to avoid ambiguity
    return {"|".join(node_types[i:i+n]) for i in range(len(node_types) - n + 1)}


def minhash_signature_from_shingles(shingle_set: set[str], num_perm: int = 128, seed0: int = 1337):
    if not shingle_set:
        return [I64_MAX] * num_perm
    sig = [I64_MAX] * num_perm
    for s in shingle_set:
        for i in range(num_perm):
            hv = _h63(s, seed0 + i)
            if hv < sig[i]:
                sig[i] = hv
    return sig

def minhash_from_node_types(
    node_types: List[str],
    shingle_n: int = 5,
    num_perm: int = 128,
    seed0: int = 1337,
) -> List[int]:
    S = shingles_from_node_types(node_types, n=shingle_n)
    return minhash_signature_from_shingles(S, num_perm=num_perm, seed0=seed0)
```

```python
import hashlib
from itertools import combinations
import ray

def band_key(sig, band_idx, r):
    start = band_idx * r
    b = b"".join(int(v).to_bytes(8, "little", signed=True) for v in sig[start:start+r])
    h = hashlib.blake2b(b, digest_size=8).hexdigest()
    return f"{band_idx}:{h}"

def lsh_pairs_map_batches(batch, k=128, b=32, max_bucket_size=2000):
    # batch is a dict-of-lists in "default" format
    ids = batch["id"]
    sigs = batch["sig"]
    ok = batch.get("sig_ok", [True]*len(ids))

    r = k // b
    buckets = {}  # band_key -> list of ids

    for _id, sig, good in zip(ids, sigs, ok):
        if not good or not sig or len(sig) != k:
            continue
        for band_idx in range(b):
            key = band_key(sig, band_idx, r)
            buckets.setdefault(key, []).append(_id)

    out_id1, out_id2 = [], []
    for key, bucket_ids in buckets.items():
        bucket_ids = sorted(set(bucket_ids))
        if len(bucket_ids) < 2 or len(bucket_ids) > max_bucket_size:
            continue
        for a, c in combinations(bucket_ids, 2):
            out_id1.append(a)
            out_id2.append(c)

    return {"id1": out_id1, "id2": out_id2}

# usage:
# ds_sig must have columns: id, sig, sig_ok
pairs_ds = ds_sig.select_columns(["id", "sig", "sig_ok"]).map_batches(
    lsh_pairs_map_batches,
    batch_format="default",
    batch_size=512,  # tweak
    fn_kwargs={"k": 128, "b": 32, "max_bucket_size": 2000},
)

# Deduplicate candidate pairs
pairs_ds = pairs_ds.groupby(["id1", "id2"]).count().drop_columns(["count()"])

pairs = pairs_ds.take_all()  # OK for small local tests
print("pairs:", len(pairs))
print(pairs[:5])
```

```python
import hashlib

def stable_cluster_id(root: str) -> str:
    # stable, readable cluster id
    return hashlib.md5(root.encode("utf-8", "ignore")).hexdigest()

class UnionFind:
    def __init__(self):
        self.parent = {}

    def find(self, x):
        if x not in self.parent:
            self.parent[x] = x
            return x
        while self.parent[x] != x:
            self.parent[x] = self.parent[self.parent[x]]
            x = self.parent[x]
        return x

    def union(self, a, b):
        ra, rb = self.find(a), self.find(b)
        if ra != rb:
            self.parent[rb] = ra

def cluster_map_from_pairs(pairs: list[dict]) -> dict[str, str]:
    """
    pairs: [{"id1":..., "id2":...}, ...]
    returns: id -> cluster_id
    """
    uf = UnionFind()
    for p in pairs:
        uf.union(p["id1"], p["id2"])

    # build id -> root
    id_to_root = {}
    for p in pairs:
        for x in (p["id1"], p["id2"]):
            id_to_root[x] = uf.find(x)

    # convert root -> stable cluster id
    root_to_cid = {}
    for root in set(id_to_root.values()):
        root_to_cid[root] = stable_cluster_id(root)

    return {x: root_to_cid[r] for x, r in id_to_root.items()}
# Collect candidate pairs locally (OK for small tests)
pairs = pairs_ds.take_all()

# Build mapping id -> cluster_id
id_to_cluster = cluster_map_from_pairs(pairs)

def add_cluster_id(row):
    _id = row["id"]
    # If the row never appeared in any pair, it’s a singleton cluster.
    row["cluster_id"] = id_to_cluster.get(_id, stable_cluster_id(_id))
    return row

ds_with_cluster = ds_sig.map(add_cluster_id)

print(ds_with_cluster.select_columns(["id", "cluster_id"]).take(5))
```

```python
import os
import hashlib
import ray

# ---------- helpers ----------
def _stable_md5(s: str) -> str:
    return hashlib.md5(s.encode("utf-8", "ignore")).hexdigest()

def _rank_fields(row):
    # Higher is "better" (we'll pick max)
    code_len = len(row.get("code_ref") or "")
    instr_len = len(row.get("instruction") or "")
    # stable tiebreak (lower hash wins, so negate by using reversed sort key later)
    id_hash = _stable_md5(row.get("id") or "")
    return code_len, instr_len, id_hash

def _pick_better(a, b):
    # Deterministic representative choice (no quality scoring):
    # 1) longest code_ref
    # 2) then longest instruction
    # 3) then smallest md5(id)
    ac, ai, ah = _rank_fields(a)
    bc, bi, bh = _rank_fields(b)

    if ac != bc:
        return a if ac > bc else b
    if ai != bi:
        return a if ai > bi else b
    return a if ah < bh else b


# ---------- Path A: exact dedup snapshot (requires canon_hash column) ----------
def write_snapshot_exact(ds, out_dir: str, snapshot_id: str, key_col: str = "canon_hash"):
    """
    Exact dedup snapshot:
      - Groups by key_col (e.g., canon_hash)
      - Keeps 1 representative per key deterministically
    Assumes ds contains key_col and original columns.
    """
    os.makedirs(out_dir, exist_ok=True)

    # Keep only rows with a non-empty key
    ds_k = ds.filter(lambda r: bool(r.get(key_col)))

    # Reduce: one row per hash key
    reps = ds_k.groupby(key_col).reduce(_pick_better)

    # Write
    path = os.path.join(out_dir, f"snapshot_exact_{snapshot_id}")
    reps.write_parquet(path)
    return reps, path


# ---------- Path B: cluster dedup snapshot (requires cluster_id column) ----------
def write_snapshot_cluster(ds, out_dir: str, snapshot_id: str, cluster_col: str = "cluster_id"):
    """
    Cluster-based snapshot (near-dup grouping already done elsewhere):
      - Groups by cluster_col
      - Keeps 1 representative per cluster deterministically
    Assumes ds contains cluster_col and original columns.
    """
    os.makedirs(out_dir, exist_ok=True)

    ds_c = ds.filter(lambda r: r.get(cluster_col) is not None and r.get(cluster_col) != "")

    reps = ds_c.groupby(cluster_col).reduce(_pick_better)

    path = os.path.join(out_dir, f"snapshot_cluster_{snapshot_id}")
    reps.write_parquet(path)
    return reps, path
```
